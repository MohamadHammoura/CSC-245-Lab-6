```python
#1
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("company_sales_data.csv")

month_numbers = data['month_number']
total_profit = data['total_profit']

plt.figure(figsize=(10, 6))
plt.plot(month_numbers, total_profit, marker='o', color='b', linestyle='-')

plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.title('Total Profit of All Months')

plt.grid(True)
plt.show()
```

```python
#2
data = pd.read_csv("company_sales_data.csv")
total_profit = data['total_profit'].sum()

plt.figure(figsize=(10, 6))
plt.plot(data['month_number'], data['total_profit'], linestyle=':', color='red', marker='o', markerfacecolor='red', linewidth=3, label=f'Total Profit: ${total_profit:,}')

plt.xlabel('Month Number')
plt.ylabel('Sold units number')
plt.title('Total Profit of All Months')

plt.legend(loc='lower right')

plt.grid(True)
plt.show()
```

```python
#3
data = pd.read_csv("company_sales_data.csv")
products = ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']

plt.figure(figsize=(10, 6))
for product in products:
    plt.plot(data['month_number'], data[product], marker='o', label=product)

plt.xlabel('Month Number')
plt.ylabel('Number of Units Sold')
plt.title('Product Sales Data')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#4
data = pd.read_csv("company_sales_data.csv")

plt.figure(figsize=(8, 6))
plt.scatter(data['month_number'], data['toothpaste'], color='b')

plt.xlabel('Month Number')
plt.ylabel('Toothpaste Sales')
plt.title('Toothpaste Sales Data')
plt.grid(True, linestyle='--')
plt.show()
```

```python
#5
data = pd.read_csv("company_sales_data.csv")

plt.figure(figsize=(10, 6))

plt.bar(data['month_number'] - 0.2, data['facecream'], width=0.4, color='b', label='Face Cream')
plt.bar(data['month_number'] + 0.2, data['facewash'], width=0.4, color='g', label='Face Wash')

plt.xlabel('Month Number')
plt.ylabel('Number of Units Sold')
plt.title('Face Cream and Face Wash Sales Data')
plt.legend()
plt.grid(True)
plt.xticks(data['month_number'])
plt.show()
```

```python
#6
data = pd.read_csv("company_sales_data.csv")

plt.figure(figsize=(10, 6))
plt.bar(data['month_number'], data['bathingsoap'], color='orange')

plt.xlabel('Month Number')
plt.ylabel('Bathing Soap Sales')
plt.title('Bathing Soap Sales Data')
plt.grid(True)
plt.savefig('bathing_soap_sales.png')  # Save the plot to hard disk
plt.show()
```

```python
#7
data = pd.read_csv("company_sales_data.csv")

plt.figure(figsize=(10, 6))
plt.hist(data['total_profit'], bins=10, color='skyblue', edgecolor='black')

plt.xlabel('Total Profit')
plt.ylabel('Frequency')
plt.title('Distribution of Total Profit')
plt.grid(True)
plt.show()
```

```python
#8
data = pd.read_csv("company_sales_data.csv")

last_year_data = data[data['month_number'] > 6]  # Considering last 6 months as last year

total_sales_last_year = last_year_data[products].sum()

plt.figure(figsize=(8, 8))
plt.pie(total_sales_last_year, labels=products, autopct='%1.1f%%', startangle=140)

plt.title('Total Sales Last Year by Product')
plt.show()
```

```python
#9
data = pd.read_csv("company_sales_data.csv")

fig, axs = plt.subplots(2, figsize=(10, 8))

axs[0].plot(data['month_number'], data['bathingsoap'], marker='o', color='b')
axs[0].set_title('Bathing Soap Sales')
axs[0].set_xlabel('Month Number')
axs[0].set_ylabel('Units Sold')

axs[1].plot(data['month_number'], data['facewash'], marker='o', color='g')
axs[1].set_title('Facewash Sales')
axs[1].set_xlabel('Month Number')
axs[1].set_ylabel('Units Sold')

plt.tight_layout()
plt.show()
```

```python
#10
data = pd.read_csv("company_sales_data.csv")

plt.figure(figsize=(10, 6))

products = ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']
sales_data = [data[product] for product in products]

plt.stackplot(data['month_number'], sales_data, labels=products)

plt.xlabel('Month Number')
plt.ylabel('Total Sales')
plt.title('Product Sales Data')
plt.legend(loc='upper left')
plt.grid(True)
plt.show()
```
